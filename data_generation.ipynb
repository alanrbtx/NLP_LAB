````python
from tqdm import tqdm
import openai 
import json
from datasets import load_dataset, Dataset
import pandas as pd

openai.api_base = "hidden"
openai.api_key = "dummy"

themes = [
    "Столицы мира",
    "Известные писатели и их произведения",
    "Исторические события",
    "Знаменитые киногерои",
    "Географические названия",
    "Спортивные достижения",
    "Зоология",
    "Физика",
    "Химия",
    "Искусство",
    "Фотография",
    "Космос",
    "Технологии",
    "Кулинария",
    "Музыка",
    "Иностранные языки",
    "Литература",
    "Игры",
    "История искусства",
    "Математика",
    "Психология",
    "Биология",
    "Анатомия",
    "Философия",
    "Политика",
    "Социология",
    "Экономика",
    "Театр",
    "Живопись",
    "Архитектура",
    "Мода",
    "История музыки",
    "Танцы",
    "Астрономия",
    "Геология",
    "Ботаника",
    "История",
    "Культурология",
    "Античность",
    "Средние века",
    "Великие открытия",
    "Великие изобретения",
    "Мировые религии",
    "Мифология",
    "Богатства мира",
    "Удивительные факты",
    "Олимпийские игры",
    "Города мира",
    "Страны мира",
    "Горные вершины",
    "Библейские сюжеты",
    "Гастрономия",
    "Туризм",
    "Животный мир",
    "Птицы",
    "Морские обитатели",
    "Диковинные растения",
    "Угрозы окружающей среде",
    "Энергетика",
    "Ветеринария",
    "Медицина",
    "Авиация",
    "Космонавтика",
    "Транспорт",
    "Автомобили",
    "История авиации",
    "История железных дорог",
    "Промышленность",
    "Инновации",
    "Стартапы",
    "Гении и изобретатели",
    "Хакерство и кибербезопасность",
    "Медицинские открытия",
    "Игры разума",
    "Открытия Великих мореплавателей",
    "Герои детских книг",
    "Известные картинки",
    "Исторические личности",
    "Достижения в науке",
    "Интересный исторический факт",
    "Тайны мира",
    "Профессии",
    "Предметы",
    "Секреты природы",
    "Четыре стихии",
    "Интересные факты о животных",
    "Игры разума",
    "Античность",
    "Мифы и легенды",
    "Чудеса света",
    "Загадки Вселенной",
    "Загадки времени",
    "Теории заговора",
    "Семь чудес света",
    "Загадки истории",
    "Популярные выражения",
    "Этимология слов",
    "Кулинарные рецепты",
    "Спортивные рекорды",
    "Головоломки и ребусы",
    "Символика и эмблемы",
    "Достижения человечества",
    "Недооцененные факты"
]


examples = []

for theme in tqdm(themes):
    prompt = f"""
    Придумай список из 5 полезных задач для вопросно- ответных систем

    Пожалуйста, следуй этим правилам:

    1. Задачи должны охватывать тему '{theme}'
    2. Ответ должен быть в формате словаря python со следующими ключами:
        1. “context”: контекст, содержащий ответ на вопрос
        2. “question”: вопрос по заданному контексту и должен быть фрагментом из context 
        3. “text”: ответ на вопрос по заданному контексту

    Пример задачи для вопросно- ответной системы:

    'context': 'Первые упоминания о строении человеческого тела встречаются в Древнем Египте. В XXVII веке до н. э. египетский врач Имхотеп описал некоторые органы и их функции, в частности головной мозг, деятельность сердца, распространение крови по сосудам. В древнекитайской книге Нейцзин (XI—VII вв. до н. э.) упоминаются сердце, печень, лёгкие и другие органы тела человека. В индийской книге Аюрведа ( Знание жизни , IX-III вв. до н. э.) содержится большой объём анатомических данных о мышцах, нервах, типах телосложения и темперамента, головном и спинном мозге.',
    'question': 'Где встречаются первые упоминания о строении человеческого тела?',
    'answers': 'в Древнем Египте'

    Выходные данные всегда должны быть только в формате словаря python, не объясняйте свои действия и не выводите ничего другого. Проявляйте творческий подход!
    """
    response = openai.ChatCompletion.create(
                    model="hidden",
                    temperature=0.0,
                    messages=[
                        {"role": "system", "content": ""},
                        {"role": "user", "content": prompt}
                    ]
                )
    answer = response["choices"][0]["message"]["content"]
    examples.append(answer)
````

### PROCESS DATA

````python
processed_data = []
for data in tqdm(raw_data[:90]):
    if "```" in data:
        splited_data = data.split("```")
        processed_data.append(splited_data[1])
        processed_data.append(splited_data[3])
        processed_data.append(splited_data[5])
        processed_data.append(splited_data[7])
        processed_data.append(splited_data[9])
    if "\n\n" in data and "```" not in data:
        splited_data = data.split("\n\n")
        processed_data.append(splited_data[1])
        processed_data.append(splited_data[3])
        processed_data.append(splited_data[5])
        processed_data.append(splited_data[7])
        processed_data.append(splited_data[9])

print(processed_data)
````

````python
processed_data
````

````python
json_data = []
for data in tqdm(processed_data):
    print(data)
    if "{" in data and "}" in data:
        text = data.replace("'", '')
        try:
            data = json.loads(text)
            json_data.append(data)
        except:
            continue
````

````python
context = []
question = []
text = []
for data in json_data:
    context.append(data['context'])
    question.append(data['question'])
    try:
        text.append(str(data['text']))
    except:
        text.append(str(data['answer']))

df = pd.DataFrame({'context': context, 'question': question, 'answers': text})
````
